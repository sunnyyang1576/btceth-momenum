```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Test 1

```python
df = pd.read_excel("data/usdt_pairs_update.xlsx")
df["time"] = pd.to_datetime(df["time"])
df = df[df["ticker"].isin(["BTCUSDT","ETHUSDT"])]
wide_df = df.pivot_table(columns=["ticker"],index=["time"],values=["close"])
```

```python
signal_df = wide_df.pct_change(1).shift(1)
ret_df = wide_df.pct_change(1)
```

```python
signal_df.columns = ["BTCUSDT","ETHUSDT"]
ret_df.columns = ["BTCUSDT","ETHUSDT"]
```

```python
signal = signal_df["BTCUSDT"]>signal_df["ETHUSDT"]
port_ret = ret_df["ETHUSDT"].copy()
port_ret[signal] = ret_df["BTCUSDT"][signal]
portfolio_ret_cum = (port_ret+1).cumprod()
```

```python
result = pd.concat([port_ret,ret_df],axis=1)
result.columns = ["portfolio_ret","BTCUSDT","ETHUSDT"]
```

```python
result_cum = (result + 1).cumprod()
```

```python
result_cum.plot(figsize=(20,10))
plt.axvline("2021-01-01",color="red")
```
